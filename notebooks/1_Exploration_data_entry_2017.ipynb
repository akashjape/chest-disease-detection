# CHEST X-RAY DATASET EXPLORATION

```python
import pandas as pd
import numpy as np

print("CHEST X-RAY DATASET EXPLORATION")
```

```python
print("\nSTEP 1: Loading Data_Entry_2017.csv")

try:
    # Loading the main data file
    df = pd.read_csv('Data_Entry_2017.csv')
    print("Successfully loaded data!")
    print(f"Dataset shape: {df.shape}")  
   
except FileNotFoundError:
    print("ERROR: Data_Entry_2017.csv not found!")
    print("Make sure the file is in the same folder as your script")
    exit()
```

```python
print("\n STEP 2: Basic Dataset Information")
print("-" * 40)

# first few rows to understand the structure
print("\nFirst 5 rows of the dataset:")
print(df.head())

print(f"\n Dataset Dimensions:")
print(f"  Total number of X-ray images: {len(df):,}")
print(f"  Number of columns/features: {df.shape[1]}")

print(f"\n Column Names:")
df.columns
```

```python
print("\n STEP 3: Disease Analysis")
print("-" * 40)

# different types of findings
print("\n Most common findings:")
finding_counts = df['Finding Labels'].value_counts()
print(finding_counts.head(10))

print(f"\n Total unique finding combinations: {df['Finding Labels'].nunique()}")

# individual diseases
print("\n Individual Disease Analysis:")

# unique diseases 
all_diseases = set()
for findings in df['Finding Labels']:
    diseases = findings.split('|')
    all_diseases.update(diseases)

print(f"   Total unique diseases: {len(all_diseases)}")
print(f"   Disease names: {', '.join(sorted(all_diseases))}")
```
